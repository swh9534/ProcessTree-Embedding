# Tokenizing

```python
import re

def tokenize_tree(tree):
    # 특수 문자를 이스케이프하여 정규 표현식 적용 + 괄호는 넣기
    special_characters = r"[\*\+\./:;<=\?\[\]\^_`{|}~\(\)]"
    tokenized_tree = re.findall(r"\b\w+\b|" + special_characters + r"|\u2192", str(tree).replace("->", "\u2192"))
    return tokenized_tree

# Tokenize each group
tokenized_processtree_naive = [tokenize_tree(tree) for tree in processtree_naive]
tokenized_processtree_simple = [tokenize_tree(tree) for tree in processtree_simple]
tokenized_processtree_concise = [tokenize_tree(tree) for tree in processtree_concise]
tokenized_processtree_straightforward = [tokenize_tree(tree) for tree in processtree_straightforward]
tokenized_processtree_complex = [tokenize_tree(tree) for tree in processtree_complex]
tokenized_processtree_spaghetti = [tokenize_tree(tree) for tree in processtree_spaghetti]
```

# Bag-of-Words

```python
def bag_of_words(tokenized_processes, global_vocab):
    vocab_list = sorted(list(global_vocab))
    bows = []

    for tokenized_process in tokenized_processes:
        process_vocab = {word: index for index, word in enumerate(vocab_list)}
        process_bow = np.zeros(len(vocab_list), dtype=int)

        for word in tokenized_process:
            word_index = process_vocab.get(word)
            process_bow[word_index] += 1

        bows.append(process_bow)

    return bows
```

```python
# 각 그룹에 대해 벡터화된 표현 생성
bows_naive = bag_of_words(tokenized_processtree_naive, global_vocab)
bows_simple = bag_of_words(tokenized_processtree_simple, global_vocab)
bows_concise = bag_of_words(tokenized_processtree_concise, global_vocab)
bows_straightforward = bag_of_words(tokenized_processtree_straightforward, global_vocab)
bows_complex = bag_of_words(tokenized_processtree_complex, global_vocab)
bows_spaghetti = bag_of_words(tokenized_processtree_spaghetti, global_vocab)
```

```python
def sentence_embedding(sentence_tokens, model):
    # Initialize an empty array to store word vectors
    word_vectors = []

    # Iterate through the tokens in the sentence
    for token in sentence_tokens:
        if token in model.wv:
            # Keep the original 64-dimensional vector for each word
            word_vector = model.wv[token]
            word_vectors.append(word_vector)

    sentence_embedding = np.mean(word_vectors, axis=0, keepdims=False)
    return sentence_embedding
```

```python
def sentence_embedding(tokens, model, embedding_dim=64):
    size = len(tokens)
    matrix = np.zeros((size, embedding_dim))
    word_table = word_dict
    for i, token in enumerate(tokens):
        vector = np.array([
            word_table[t] for t in token
            if t in word_table
        ])

        if vector.size != 0:
            final_vector = np.mean(vector, axis=0)
            matrix[i] = final_vector

    sentence_embedding = np.mean(matrix, axis=0)

    return sentence_embedding
```

## BERT

```python
def embed_processes_with_bert(tokenized_processes):

    # 임베딩을 저장할 리스트 초기화
    bert_embeddings = []

    # tokenized_processes에 대한 반복문
    for tree_tokens in tokenized_processes:
        # 프로세스 트리를 텍스트로 펼치고 토큰화
        process_tree_text = " ".join(tree_tokens)
        tokens = bert_tokenizer(process_tree_text, return_tensors='pt').to(device)

        # BERT 모델에 입력
        with torch.no_grad():
            output = bert_model(**tokens)

        # [CLS] 토큰의 임베딩 사용
        cls_embedding = output.last_hidden_state[:, 0, :].cpu().squeeze().numpy()

        bert_embeddings.append(cls_embedding)

    # 모든 프로세스 트리에 대한 임베딩이 bert_embeddings에 저장됨
    return bert_embeddings
```

```python
import random
import string

def mask_data(sentence_tokens, mask_ratio=0.15):
    masked_tokens = list(sentence_tokens)

    # Identify the lowercase alphabet tokens in the sentence
    alphabet_tokens = [token for token in sentence_tokens if token.isalpha() and token.islower()]

    # Randomly select lowercase alphabet tokens to mask
    tokens_to_mask = random.sample(alphabet_tokens, int(len(alphabet_tokens) * mask_ratio))

    for token in tokens_to_mask:
        masked_tokens[masked_tokens.index(token)] = "[MASK]"

    return masked_tokens

def replace_data(sentence_tokens, replace_ratio=0.15, vocab_size=10000):
    replaced_tokens = list(sentence_tokens)

    # Identify the lowercase alphabet tokens in the sentence
    alphabet_tokens = [token for token in sentence_tokens if token.isalpha() and token.islower()]

    # Randomly select lowercase alphabet tokens to replace
    tokens_to_replace = random.sample(alphabet_tokens, int(len(alphabet_tokens) * replace_ratio))

    # Generate a vocabulary of random lowercase alphabet tokens
    random_vocab = [random.choice(string.ascii_lowercase) for _ in range(vocab_size)]

    for token in tokens_to_replace:
        # Replace with a new lowercase token from the given vocabulary
        new_token = random.choice(random_vocab)
        replaced_tokens[replaced_tokens.index(token)] = new_token

    return replaced_tokens
```

# Levenshtein(edit) distance-based similarity among Linear Process Trees (Baseline)

```python
import Levenshtein

def calculate_levenshtein_distance(tree1, tree2):
    str_tree1 = str(tree1).replace(" ", "").replace("'", "").replace('tau', '0')
    str_tree2 = str(tree2).replace(" ", "").replace("'", "").replace('tau', '0')
    distance = Levenshtein.distance(str_tree1, str_tree2)
    return distance

def calculate_similarity(tree1, tree2):
    max_distance = max(len(str(tree1)), len(str(tree2)))
    distance = calculate_levenshtein_distance(tree1, tree2)
    similarity = 1 - (distance / max_distance)
    return similarity

def calculate_average_similarity(tree_group1, tree_group2):
    total_similarity = 0.0

    for tree1 in tree_group1:
        for tree2 in tree_group2:
            similarity = calculate_similarity(tree1, tree2)
            total_similarity += similarity

    # 프로세스 트리 쌍의 총 수
    total_pairs = len(tree_group1) * len(tree_group2)

    # 전체 유사도의 평균
    average_similarity = total_similarity / total_pairs

    return average_similarity
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.children = []

    def add_child(self, child):
        self.children.append(child)

    def print_tree(self, depth=0):
        print("  " * depth + self.value)
        for child in self.children:
            child.print_tree(depth + 1)
    def count_nodes(self):
        # 현재 노드를 포함하여 노드 수 계산
        count = 1
        for child in self.children:
            count += child.count_nodes()
        return count

    def count_edges(self):
        # 현재 노드의 자식 수를 엣지로 간주하여 엣지 수 계산
        edge_count = len(self.children)
        for child in self.children:
            edge_count += child.count_edges()
        return edge_count
```

```python
def tree_edit_distance(A, B):
    # 노드 A를 루트로 하는 트리와 노드 B를 루트로 하는 트리 사이의 최소 편집 거리를 계산
    def min_distance(root1, root2, memo):
        if (root1, root2) in memo:
            return memo[(root1, root2)]
        elif root1 is None and root2 is None:
            return 0
        elif root1 is None:
            return 1 + sum(min_distance(None, child, memo) for child in root2.children)
        elif root2 is None:
            return 1 + sum(min_distance(child, None, memo) for child in root1.children)
        else:
            cost = 0 if root1.value == root2.value else 1
            
            # 두 노드의 자식들을 매칭
            from itertools import permutations
            children1 = root1.children
            children2 = root2.children
            size1, size2 = len(children1), len(children2)
            min_cost = float('inf')

            # 모든 가능한 자식 노드들의 순열에 대해 편집 거리를 계산
            for perm in permutations(children2):
                # 자식들 간 비용 계산
                current_cost = sum(min_distance(child1, child2, memo) for child1, child2 in zip(children1, perm))
                current_cost += abs(size1 - size2)  # 차이나는 자식 수만큼 추가 또는 삭제 비용
                min_cost = min(min_cost, current_cost)
                
            memo[(root1, root2)] = cost + min_cost
            return memo[(root1, root2)]

    # 메모이제이션을 위한 딕셔너리
    memo = {}
    return min_distance(A, B, memo)
```

```python
def text_to_tree(sentence):
    stack = []
    root = TreeNode(None)
    sentence = sentence.replace(" ", "")
    for i in range(len(sentence)):
        if sentence[i] != '(' and sentence[i] != ')':
            cur = sentence[i]
            child = TreeNode(cur)
            root.add_child(child)
        elif sentence[i] == '(':
            stack.append(child)
            root = child
        else:
            stack.pop()
            if stack != []:
                root = stack[-1]
            

    return root
```

```python
def max_edit_distance(graph1, graph2):
    # 그래프의 노드 수와 엣지 수 계산
    node_count1 = graph1.count_nodes()
    edge_count1 = graph1.count_edges()
    node_count2 = graph2.count_nodes()
    edge_count2 = graph2.count_edges()

    # 두 그래프의 노드 수와 엣지 수 중에서 더 큰 값을 최대 편집 거리로 지정
    max_node_count = max(node_count1, node_count2)
    max_edge_count = max(edge_count1, edge_count2)

    return max_node_count + max_edge_count
```

```python
def graph_edit_similarity(sentence1, sentence2):
    graph1 = text_to_tree(sentence1)
    graph2 = text_to_tree(sentence2)
    
    ged = tree_edit_distance(graph1, graph2)
    maxged = max_edit_distance(graph1, graph2)
    print(ged)
    print(maxged)
    sim = 1 - (ged / maxged)
    print(sim)
```

## cosine 유사도 함수 정의

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def cosine_avg(embeddings_groups):
    num_groups = len(embeddings_groups)
    similarity_matrix = np.zeros((num_groups, num_groups))

    for i in range(num_groups):
        for j in range(num_groups):
            embeddings1 = embeddings_groups[i]
            embeddings2 = embeddings_groups[j]

            similarity_matrix[i, j] = np.mean(cosine_similarity(embeddings1, embeddings2))

    # Print the similarity matrix
    print("Similarity Matrix:")
    print(similarity_matrix)
```
